This notebook contains a few examples of DRN SOM neurons with bursty firing patterns and/or perithreshold inward currents. All cells shown here were positively identified using fluorescence imaging. The location of each cell within the DRN is sketched in my lab book; according to this, all cells shown here were at or below the lower diagonal from the aqueduct.

## Load data

```python
DATA_PATH = os.path.join('data', 'raw', 'GABA', 'localized_cells')

def adp(data_path, fnames):
    """Construct list of fnames with data_path pre-pended to each fname."""
    return [os.path.join(data_path, fname) for fname in fnames]

DRN417 = Cell(
    name = 'DRN417-SOM-lateral', # Shown below diagonal in lab book sketch. Images 110119-DRN417wide/narrow.tiff.
    IA_steps_TTX = adp(DATA_PATH, ['19111064.abf', '19111065.abf']),
    V_steps = adp(DATA_PATH, ['19111062.abf', '19111063.abf'])
)

DRN419 = Cell(
    name = 'DRN419-SOM-lateral', # Shown below diagonal in lab book sketch.
    IA_steps = adp(DATA_PATH, 
                   ['19111090.abf', '19111091.abf', '19111092.abf', 
                    '19111093.abf', '19111094.abf', '19111095.abf']),
    AEC = adp(DATA_PATH, ['19111081.abf', '19111084.abf', '19111087.abf']),
    train = adp(DATA_PATH, ['19111082.abf', '19111085.abf', '19111088.abf']),
    test = adp(DATA_PATH, ['19111083.abf', '19111086.abf', '19111089.abf'])
)

DRN420 = Cell(
    name = 'DRN420-SOM-medial', # Shown on midline in lab book sketch.
    V_steps = adp(DATA_PATH, ['19111102.abf'])
)

DRN421 = Cell(
    name = 'DRN421-SOM-medial', # Shown on midline in lab book sketch.
    IA_steps = adp(DATA_PATH, ['19114000.abf', '19114001.abf', '19114002.abf', '19114003.abf', '19114004.abf']),
    I_steps = adp(DATA_PATH, ['19114005.abf']),
    long_I_steps = adp(DATA_PATH, ['19114006.abf']),
    AEC = adp(DATA_PATH, ['19114007.abf', '19114010.abf', '19114013.abf']),
    train = adp(DATA_PATH, ['19114008.abf', '19114011.abf', '19114014.abf']),
    test = adp(DATA_PATH, ['19114009.abf', '19114012.abf', '19114015.abf'])
)
```
